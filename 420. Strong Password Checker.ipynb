```python
class Solution:
    def strongPasswordChecker(self, password: str) -> int:
        
        rpt = []
        it = 0
        cc = 0
        is_low = False
        is_up = False
        is_digit = False
        unq = 0
        for i in range(len(password)):
            if not is_low and password[i] >= 'a' and password[i] <= 'z':
                unq += 1
                is_low = True  
            if not is_up and password[i] >= 'A' and password[i] <= 'Z':
                unq += 1
                is_up = True  
            if not is_digit and password[i] >= '0' and password[i] <= '9':
                unq += 1
                is_digit = True  
            if password[i] != cc:
                if it >= 3:
                    rpt.append(it)
                cc = password[i]
                it = 1
            else:
                it += 1
        if it >= 3:
            rpt.append(it)
        
        
        if (len(rpt) == 0 and unq == 3 and len(password) >= 6 and len(password) <= 20):
            return 0
        
                          
        if len(password) <= 20:
            return max(sum([i // 3 for i in rpt]), 3 - unq, 6 - len(password), 0)


        
        
        
        need_del = len(password) - 20
        rpt.sort()
        
        plus_add = 0
        while True:
            if need_del >= sum([i - 2 for i in rpt]):
                return need_del + max(0, 3 - unq) + plus_add
            if sum([1 for i in rpt if i == 5]) >= 1:
                i = sum([1 for i in rpt if i < 5])
                del rpt[i]
                unq += 1
                plus_add += 1
                if need_del >= sum([i - 2 for i in rpt]):
                    return need_del + max(0, 3 - unq) + plus_add

            elif rpt[len(rpt) - 1] > 5:
                i = sum([1 for i in rpt if i < 5])
                rpt[i] -= 3
                plus_add += 1
                if rpt[i] < 3:
                    del rpt[i]
                else:
                    rpt.sort()
                unq += 1

            elif rpt[0] == 3 and need_del > 0:
                plus_add += 1
                del rpt[0]
                need_del -= 1
            elif rpt[0] == 4 and need_del >= 2:
                plus_add += 2
                del rpt[0]
                need_del -= 2
            elif rpt[0] == 4 and need_del == 1:
                plus_add += 2
                del rpt[0]
                need_del = 0
            elif need_del == 0:
                return max(0, 3 - unq) + plus_add + sum([i // 3 for i in rpt])
            
```
