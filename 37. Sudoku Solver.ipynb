```python
class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
            self.fnl = [[0]*9 for i in range(9)]
            self.notmay = [[set() for j in range(9)] for i in range(9)]
            self.need_print = False
            self.in_qst = 0

            for rr in range(9):
                for cc in range(9):
                    if board[rr][cc] != '.':
                        self.fnl[rr][cc] = int(board[rr][cc])
                        self.defind(rr, cc, int(board[rr][cc]))
            self.print_state()
            if self.need_print:
                print('Start solution')
            
            self.find_simple_sol()               
            
            # Start trye
            if self.in_qst < 81:
                self.try_one_sample()
    



            self.print_state('Result ' + ' ' + str(self.check_valid()) + str(self.in_qst))
#             print(self.check_valid(), self.in_qst, '\n\n\n')
            for rr in range(9):
                for cc in range(9):
                    board[rr][cc] = str(self.fnl[rr][cc])
    def print_state(self, xx = ''):
        if not(self.need_print):
            return False
        if (xx != ''):
            print(xx)
        print('------------------------------')
        for rr in range(9):
            for cc in range(9):
                if self.fnl[rr][cc] != 0:
                    print(' ', self.fnl[rr][cc], ' ', end='', sep = '')
                else:
                    print(' - ', end='')
                if cc % 3 == 2:
                    print('|', end = '')
            print()
            if rr % 3 == 2:
                print('------------------------------')
    def defind(self,  rr, cc, k):
        self.fnl[rr][cc] = k
        self.in_qst += 1
        for xx in range(9):
            if self.fnl[xx][cc] == 0:
                self.notmay[xx][cc].add(k)
            if self.fnl[rr][xx] == 0:
                self.notmay[rr][xx].add(k)
        for r2 in range((rr // 3) * 3, (rr // 3) * 3 + 3):        
            for c2 in range((cc // 3) * 3, (cc // 3) * 3 + 3):        
                if self.fnl[r2][c2] == 0:
                    self.notmay[r2][c2].add(k)
    def try_only_ony(self):
        for rr in range(9):
            for cc in range(9):
                if self.fnl[rr][cc] == 0 and len(self.notmay[rr][cc]) == 8:
                    k = [i for i in range(1, 10) if not(i in self.notmay[rr][cc])][0]
                    self.defind(rr, cc, k)
            
    def try_only_ony_cell(self):
        # if stay
        for rr in range(9):        
            for cc in range(9):
                if self.fnl[rr][cc] == 0:
                    for k in [i for i in range(1, 10) if not(i in self.notmay[rr][cc])]:
                        my = 0
                        for r2 in range((rr // 3) * 3, (rr // 3) * 3 + 3):        
                            for c2 in range((cc // 3) * 3, (cc // 3) * 3 + 3):        
                                if (r2 != rr or c2 != cc) and self.fnl[r2][c2] == 0 and not(k in self.notmay[r2][c2]):
                                    my = 1
                        if my == 0:
                            self.defind(rr, cc, k)
    def check_valid(self):
        # if stay
        for rr in range(9):
            mp = set()
            cnt = 0
            for cc in range(9):
                if self.fnl[rr][cc] != 0:
                    mp.add(self.fnl[rr][cc])
                    cnt += 1
            if len(mp) != cnt:
                return False
            
        for cc in range(9):
            mp = set()
            cnt = 0
            for rr in range(9):
                if self.fnl[rr][cc] != 0:
                    mp.add(self.fnl[rr][cc])
                    cnt += 1
            if len(mp) != cnt:
                return False
            
        for rr in range(3):        
            for cc in range(3):
                mp = set()
                cnt = 0
                for r2 in range(rr * 3, rr * 3 + 3):        
                    for c2 in range(cc * 3, cc * 3 + 3):        
                        if self.fnl[r2][c2] != 0:
                            mp.add(self.fnl[r2][c2])
                            cnt += 1
            if len(mp) != cnt:
                return False
        for rr in range(9):        
            for cc in range(9):
                if self.fnl[rr][cc] == 0 and len(self.notmay[rr][cc]) == 9:
                    return False
        return True
    
    def get_variant(self):
        psb_var = []
        rr, cc = 0, 0
        while rr  < 9 and self.fnl[rr][cc] != 0:
            while cc  < 9 and self.fnl[rr][cc] != 0:
                cc += 1
            if cc >= 9:
                cc = 0
                rr += 1
        if rr >= 9:
            return []
        for xx in [i for i in range(1, 10) if not(i in self.notmay[rr][cc])]:
            psb_var.append((rr, cc, xx))
        return psb_var
    def get_snap(self):
        qst = self.in_qst
        fnl2 = [[0]*9 for i in range(9)]
        notmay2 = [[set() for j in range(9)] for i in range(9)]
        for rr in range(9):
            for cc in range(9):
                fnl2[rr][cc] = self.fnl[rr][cc]
                for i in self.notmay[rr][cc]:
                    notmay2[rr][cc].add(i)        
        return [qst, fnl2, notmay2]
    def try_one_sample(self):
        # Get variants
        psb_var = self.get_variant()
        self.print_state('Try eample ' + str(self.check_valid()) + ', qst: ' + str(self.in_qst))
        
        for rrr in psb_var:
            qst, fnl2, notmay2 = self.get_snap()
            self.defind(rrr[0], rrr[1], rrr[2])
            self.find_simple_sol()
            if self.in_qst == 81 and self.check_valid():
                return True
            
            if self.try_one_sample():
                return True
            else:
                self.in_qst, self.fnl, self.notmay = qst, fnl2, notmay2
        return False
    def find_simple_sol(self):
        lst = self.in_qst - 1
        while lst != self.in_qst and self.in_qst < 81:
            lst = self.in_qst
            self.try_only_ony()
            self.try_only_ony_cell();
```
