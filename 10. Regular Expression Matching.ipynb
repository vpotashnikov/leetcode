```python
class Solution:
    def regr(self, s: str, p: str) -> bool:
        if ((s + '#' + p) in self.alrd):
            return False
        else:
            self.alrd.add(s + '#' + p)
#         print ('- 0 -', s, p)
        while len(s) != 0 and len(p) != 0 and (len(p) <= 1 or p[1] != '*'):
            if p[0] == '.':
                p = p[1:len(p)]
                s = s[1:len(s)]
            elif p[0] != s[0]:
                return False
            else:
                p = p[1:len(p)]
                s = s[1:len(s)]
        
        
        if len (p) == 0:
            return (len(s) == 0)
        
        if len(s) == 0:
            if len(p) % 2 != 0:
                return False
            i = 0
            while i + 1 < len(p):
                if p[i + 1] != '*':
                    return False
                i += 2

            return True
            
        if len(p) >= 2 and p[1] == '*':
            if self.regr(s, p[2:len(p)]):
                return True
            
            if p[0] == '.':
                if len(p) == 2:
                    return True
                i = 0
                while i < len(s):
                    if self.regr(s[(i + 1):len(s)], p[2:len(p)]):
                        return True
                    else:
                        i += 1
                return False
            else:
                k = p[0]
                i = 0
                while i < len(s) and s[i] == k:
                    # print ('---', i, k, s, p)
                    if self.regr(s[(i + 1):len(s)], p[2:len(p)]):
                        return True
                    else:
                        i += 1                   
                        
                if i == len(s) and len(p) == 2 and s[i - 1] == k:
                    return True
                return False
        return False    
            
        
    def isMatch(self, s: str, p: str) -> bool:
        self.alrd = set()
        return self.regr(s, p)
```
